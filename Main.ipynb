-->> Here I am create a pie chart of total team member in specific team.

```python
import matplotlib.pyplot as plt

df2 = pd.read_csv('team.csv')
df2.head()


data_pie = df2['team'].value_counts().rename_axis('team').reset_index(name='team_member')

plt.figure(figsize=(10,10))
plt.pie(data_pie.team_member, labels=data_pie.team, startangle=90, autopct='%.1f%%')
plt.title("Team member")
plt.show()
```

-->> Here I am create a pie chart of all member Position..

```python
data_pie1 = df2['position'].value_counts().rename_axis('position').reset_index(name='position_a')

plt.figure(figsize=(10,10))
plt.pie(data_pie1.position_a, labels=data_pie1.position, startangle=90, autopct='%.1f%%')
plt.title("Team member with his position")
plt.show()
```

-->> Here I am create a position chart of BAL team

```python
import pandas as pd
df = pd.read_csv('team.csv')
df1 = pd.DataFrame(df.loc[df['team'] == 'BAL'])
chart = df1['position'].value_counts().rename_axis('position').reset_index(name='bal_team_member')

plt.figure(figsize=(10,10))
plt.pie(chart.bal_team_member, labels=chart.position, startangle=90, autopct='%.1f%%')
plt.title("BAL team member")
plt.show()
```
